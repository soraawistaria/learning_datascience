membaca dataset

```python
import pandas as pd

df = pd.read_csv('healthcare-dataset-stroke-data.csv')
df.head(5)
```

Mengecek total missing value di tiap kolom

```python
print(df.isnull().sum())
```

Mengisi NaN di BMI dengan median

actually, menangani missing values bisa juga dengan K-Nearest Neighbours untuk hasil yang lebih akurat

```python
df['bmi'] = df['bmi'].fillna(df['bmi'].median())

print(df.isnull().sum())
```

```python
# Kategori BMI

# under 18.5 = kurus
# 18.5 - 24  = normal
# 24 - 28 = gemuk
# above 28 = obesitas
```
