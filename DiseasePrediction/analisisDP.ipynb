```python
df.drop('Unnamed: 133', axis=1) 
#axis=0 untuk menghapus row, axis=1 untuk menghapus column

df.head()
```

```python
df = df.drop('Unnamed: 133', axis=1)

df.head()
```

```python
print(df['prognosis'].value_counts())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

symptom_counts = df.drop('prognosis', axis=1).sum().sort_values(ascending=False)

top_10 = symptom_counts.head(10)

plt.figure(figsize=(5,3))
sns.barplot(x=top_10.values, y=top_10.index, palette='viridis')
plt.title('Top 10 Disease Symptoms')
plt.xlabel('Jumlah Munculnya Gejala')
plt.show()
```
