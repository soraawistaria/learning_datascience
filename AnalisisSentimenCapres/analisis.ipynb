```python
def clean_twitter_text(text):
    text = re.sub(r'@[A-Za-z0-9_]+', '', text)
    text = re.sub(r'#\w+', '', text)
    text = re.sub(r'RT[\s]+', '', text)
    text = re.sub(r'https?://\S+', '', text)

    text = re.sub(r'[^A-Za-z0-9 ]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()

    return text
```

```python
df['full_text'] = df['full_text'].apply(clean_twitter_text)
```

```python
df['full_text'] = df['full_text'].str.lower()
```

Preprocessing

```python
#Normalisasi

norm = {" yg ": " yang ",
        " ganjar " : " ",
        " prabowo " : " ",
        "nggak": "tidak",
        "gak" : "tidak",
        "bangetdari": "banget dari",
        "bgt" : "banget",
        "jd" : "jadi",
        "tuk" : "untuk",
        "kk" : "kakak"}

def normalisasi(str_text):
    for i in norm:
        str_text = str_text.replace(i, norm[i])
    return str_text

df['full_text'] = df['full_text'].apply(lambda x: normalisasi(x))
df
```

```python
#Stopwords
import Sastrawi
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory, StopWordRemover, ArrayDictionary
more_stop_words = ["tidak"]   #buang kata yg gapunya bobot

stop_words = StopWordRemoverFactory().get_stop_words()
stop_words.extend(more_stop_words)

new_array = ArrayDictionary(stop_words)
stop_words_remover__new = StopWordRemover(new_array)

def stopword(str_text):  
    str_text = stop_words_remover__new.remove(str_text)
    return str_text

df['full_text'] = df['full_text'].apply(lambda x : stopword(x))
df.head()
```

```python
#Tokenize

tokenized = df['full_text'].apply(lambda x:x.split())  #memisahkan kata-kata
tokenized
```

```python
#Stemming
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

def stemming(text_cleaning):
    factory = StemmerFactory()
    stemmer = factory.create_stemmer()
    do = []
    for w in text_cleaning:
        dt = stemmer.stem(w)
        do.append(dt)

    d_clean = []
    d_clean = " ".join(do)
    print(d_clean)
    return d_clean

tokenized = tokenized.apply(stemming)

tokenized.to_csv('AfterProcess/ProcessingAnies.csv', index=False)
data = pd.read_csv('AfterProcess/ProcessingAnies.csv', encoding='latin1')
```

```python
import pandas as pd
data = pd.read_csv('AfterProcess/TerjemahanAnies.csv', index_col=0)
data_tweet = list(data['tweet_english'])
polaritas = 0

status = []
total_positif = total_negatif = total_netral = total = 0

for i, tweet in enumerate(data_tweet):
    analysis = TextBlob(tweet)
    polaritas += analysis.polarity

    if analysis.sentiment.polarity > 0.0:
        total_positif += 1
        status.append('Positif')
    elif analysis.sentiment.polarity == 0.0:
        total_netral += 1
        status.append('Netral')
    else:
        total_negatif += 1
        status.append('Negatif')

    total += 1

print(f'Hasil Analisis Data:\nPositif = {total_positif}\nNetral = {total_netral}\nNegatif = {total_negatif}')
print(f'\nTotal Data : {total}')
```

```python
data['klasifikasi'] = status
data
```

```python
from wordcloud import WordCloud, STOPWORDS

def plot_cloud(wordcloud):
    plt.figure(figsize=(6,4))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.show

all_words = ' '.join([tweets for tweets in data['full_text']])

wordcloud = WordCloud(
    width=3000,
    height=2000,
    random_state=3,
    background_color='black',
    colormap='Blues_r',
    collocations=False,
    stopwords=STOPWORDS
).generate(all_words)

plot_cloud(wordcloud)
```
